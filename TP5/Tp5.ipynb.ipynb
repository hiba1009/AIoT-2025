## 1. Environment Setup and Data Loading

### ğŸ“ Task 1.1: Setup and Data Loading

```python
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras import layers, models

(x_train, y_train), (x_test, y_test) = keras.datasets.fashion_mnist.load_data()
```

```python
x_train = x_train / 255.0
x_test = x_test / 255.0
```

```python
x_train_mlp = x_train.reshape((x_train.shape[0], 28*28))
x_test_mlp = x_test.reshape((x_test.shape[0], 28*28))
x_train_mlp.shape
```

```python
x_train_cnn = x_train.reshape((x_train.shape[0], 28, 28, 1))
x_test_cnn = x_test.reshape((x_test.shape[0], 28, 28, 1))
x_train_cnn.shape
```

## 2. Model Implementation

### ğŸ“ Task 2.1: Implement and Compile the MLP Model

```python
mlp_model = models.Sequential([
    layers.Input(shape=(784,)),
    layers.Dense(256, activation='relu'),
    layers.Dense(128, activation='relu'),
    layers.Dense(10, activation='softmax')
])

mlp_model.compile(optimizer='adam',
                  loss='sparse_categorical_crossentropy',
                  metrics=['accuracy'])
mlp_model.summary()
```

### ğŸ“ Task 2.2: Implement and Compile the CNN Model

```python
cnn_model = models.Sequential([
    layers.Conv2D(16, (3,3), activation='relu', input_shape=(28,28,1)),
    layers.MaxPooling2D(2,2),
    layers.Conv2D(32, (3,3), activation='relu'),
    layers.MaxPooling2D(2,2),
    layers.Flatten(),
    layers.Dense(64, activation='relu'),
    layers.Dense(10, activation='softmax')
])

cnn_model.compile(optimizer='adam',
                  loss='sparse_categorical_crossentropy',
                  metrics=['accuracy'])

cnn_model.summary()
```

## 3. Training and Evaluation

### ğŸ“ Task 3.1: Train the MLP

```python
history_mlp = mlp_model.fit(
    x_train_mlp, y_train,
    epochs=5,
    batch_size=64,
    validation_data=(x_test_mlp, y_test),
    verbose=2
)
```

### ğŸ“ Task 3.2: Train the CNN

```python
history_cnn = cnn_model.fit(
    x_train_cnn, y_train,
    epochs=5,
    batch_size=64,
    validation_data=(x_test_cnn, y_test),
    verbose=2
)
```

### ğŸ“ Task 3.3: Evaluate and Report

```python
mlp_test_loss, mlp_test_acc = mlp_model.evaluate(x_test_mlp, y_test, verbose=0)
cnn_test_loss, cnn_test_acc = cnn_model.evaluate(x_test_cnn, y_test, verbose=0)

print(f"MLP Test Accuracy: {mlp_test_acc:.4f}, Loss: {mlp_test_loss:.4f}")
print(f"CNN Test Accuracy: {cnn_test_acc:.4f}, Loss: {cnn_test_loss:.4f}")
```
